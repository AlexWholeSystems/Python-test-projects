# WSP Murder Mystery
## There has been a murder, and we need your help to find out who did it...

We know a few clues about the murderer, so using the pictures can you solve the case?

```python
suspects = ['Alex','Paul','Chris','Peter','Mark','Heather','Helen', 'James',
            'Nick', 'Sylvia', 'Claire', 'Pauline', 'Tabitha', 'Gwyn']
print(len(suspects))
glasses = {suspects[0]:False,suspects[1]:True,suspects[2]:False,
           suspects[3]:False,suspects[4]:"unknown",suspects[5]:"unknown",
           suspects[6]:True, suspects[7]:False, suspects[8]:False,
           suspects[9]:"unknown",suspects[10]:True,suspects[11]:False,
           suspects[12]:"unknown",suspects[13]:False}
role = {suspects[0]:"Analyst",suspects[1]:"Consultant",suspects[2]:"Analyst",suspects[3]:"Manager/Director",
           suspects[4]:"Consultant",suspects[5]:"Manager/Director",suspects[6]:"Consultant", suspects[7]:"Manager/Director",
           suspects[8]:"Consultant",suspects[9]:"Associate",suspects[10]:"Associate",suspects[11]:"Associate",
           suspects[12]:"Associate",suspects[13]:"Associate"}
sex = {suspects[0]:"F",suspects[1]:"M",suspects[2]:"M",suspects[3]:"M",
           suspects[4]:"M",suspects[5]:"F",suspects[6]:"F", suspects[7]:"M",
           suspects[8]:"M",suspects[9]:"F",suspects[10]:"F",suspects[11]:"F",
           suspects[12]:"F",suspects[13]:"F"}
hair = {suspects[0]:"Brown",suspects[1]:"Blonde",suspects[2]:"Brown",suspects[3]:"Brown",
           suspects[4]:"unknown",suspects[5]:"unknown",suspects[6]:"Blonde", suspects[7]:"Brown",
           suspects[8]:"Brown",suspects[9]:"unknown",suspects[10]:"Blonde",suspects[11]:"Brown",
           suspects[12]:"unknown",suspects[13]:"Brown"}

import random

highscore=0
leaderboard={}
score=60
continue_playing="Y"
while continue_playing=="Y":
    murderer=random.choice(suspects)
    name=input("please enter your name: \n")
    print(f"Hello {name}, there has been a murder and we need your help to find out who did it.")
    for x in range (5):
        guess = input("Who was the murderer? \n")   
        score-=10
        if guess==murderer:
            print(f"Congratulations, the murderer was {murderer}")
            break
        else:
            if x==0:
                print(f"Clue 1: The murderer's gender is {sex[murderer]}")
                print(f"Your score is {score}")
            if x==1:
                print(f"Clue 2: The murderer wears glasses is {glasses[murderer]}")
                print(f"Your score is {score}")
            if x==2:
                print(f"Clue 3: The murderer's hair colour is {hair[murderer]}")
                print(f"Your score is {score}")
            if x==3:
                print(f"Clue 4: The murderer's role is {role[murderer]}")
                print(f"Your score is {score}")
            if x==4:
                print("You lose")
                
    print(f"Congratulations, your score was {score}")
    if score>highscore:
        highscore=score
        print(f"Congratulations! {score} is the new highscore!")
    else:
        print(f"The highscore is {highscore}")   
              
    leaderboard[name]=score
    print(f"The leaderboard looks like this : {leaderboard}")
    
    continue_playing=input("would you like to keep playing? (Y/N)")
```
